```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_circle(center_x, center_y, radius, **params):
    # theta goes from 0 to 2pi
    theta = np.linspace(0, 2*np.pi, 100)
    # compute x1 and x2
    x = center_x + radius*np.cos(theta)
    y = center_y + radius*np.sin(theta)
    plt.plot(x, y, **params)

def plot_line_segment(start_x, start_y, end_x, end_y, **params):
    # The paremetrization of the segment
    t = np.linspace(0,1, 200)
    x = start_x * t + end_x * (1-t)
    y = start_y * t + end_y * (1-t)
    
    plt.plot(x,y, **params)

plt.figure(figsize=(15,8))
plot_circle(0,0, 1, label = r'$\alpha$-limit-set')
plot_circle(3,0, 1, label = r'$\omega$-limit-set')
plot_line_segment(1, 0, 2, 0, label = 'trajectory')
plt.legend(fontsize = 'x-large')
plt.axis('off')
plt.gca().set_aspect('equal')
plt.savefig('figures/handcuffs.pdf', bbox_inches = 'tight', dpi = 300)
plt.show()
```
